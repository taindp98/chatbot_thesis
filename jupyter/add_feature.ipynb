# Load new feature

```python
cp_priority = list_rm_2019.copy()
```

## priority

```python
new_feature = pd.read_csv('/home/taindp/PycharmProjects/thesis/data/add_feature_feb9.csv')
new_feature['major_code'] = [str(item) for item in list(new_feature['major_code'])]
new_feature['criteria'] = [str(item) for item in list(new_feature['criteria'])]
new_feature['priority_case'] = [str(item) for item in list(new_feature['priority_case'])]
new_feature['capacity_case'] = [str(item) for item in list(new_feature['capacity_case'])]
```

```python
df_priority = pd.DataFrame(cp_priority)
```

```python
list_priority_point = []
list_priority_case = []
list_priority_rm_subject_group = []
list_priority_rm_subject = []
for item in cp_priority:
    major_code = item['major_code'][0]
    type_edu = item['type_edu'][0]
#     print(type_edu)
    df_iloc = new_feature[(new_feature['major_code']==major_code) & (new_feature['type_edu']==type_edu)]
    priority_point = [df_iloc['priority_case'].iloc[0]]
    list_priority_case.append(['ưu tiên xét tuyển'])
    list_priority_rm_subject_group.append([])
    list_priority_rm_subject.append([])
    list_priority_point.append(priority_point)
```

```python
df_priority['point'] = list_priority_point
df_priority['case'] = list_priority_case
df_priority['subject'] = list_priority_rm_subject
df_priority['subject_group'] = list_priority_rm_subject_group
df_priority.head(1)
```

# capacity

```python
cp_capacity = list_rm_2019.copy()
df_capacity = pd.DataFrame(cp_capacity)
```

```python
list_capacity_point = []
list_capacity_case = []
list_capacity_rm_subject_group = []
list_capacity_rm_subject = []
for item in cp_capacity:
    major_code = item['major_code'][0]
    type_edu = item['type_edu'][0]
#     print(type_edu)
    df_iloc = new_feature[(new_feature['major_code']==major_code) & (new_feature['type_edu']==type_edu)]
    capacity_point = [df_iloc['capacity_case'].iloc[0]]
    list_capacity_case.append(['đánh giá năng lực'])
    list_capacity_rm_subject_group.append([])
    list_capacity_rm_subject.append([])
    list_capacity_point.append(capacity_point)
```

```python
df_capacity['point'] = list_capacity_point
df_capacity['case'] = list_capacity_case
df_capacity['subject'] = list_capacity_rm_subject
df_capacity['subject_group'] = list_capacity_rm_subject_group
df_capacity.head(1)
```

# Concat

```python
df_concat = pd.concat([df_query,df_priority,df_capacity],axis=0)
df_concat
```

```python
del df_concat['_id']
```

```python
list_criteria = []
dict_concat = df_concat.to_dict('records')
for item in dict_concat:
    major_code = item['major_code'][0]
    type_edu = item['type_edu'][0]
    df_iloc = new_feature[(new_feature['major_code']==major_code) & (new_feature['type_edu']==type_edu)]
    criteria = [df_iloc['criteria'].iloc[0]]
    list_criteria.append(criteria)
df_concat['criteria'] = list_criteria
```

```python
df_concat
```

```python
list_case_clean = pickle.load(open('/home/taindp/PycharmProjects/thesis/data/list_case_clean.pkl','rb'))
list_case_clean
```

```python
dict_concat = df_concat.to_dict('records')

list_object = []
list_register = []
list_admission = []
list_time = []

for item in dict_concat:
    case = item['case'][0]
    if case == 'thi tốt nghiệp trung học phổ thông':
        list_object.append(list_case_clean['formal']['object'])
        list_register.append(list_case_clean['formal']['register'])
        list_admission.append(list_case_clean['formal']['admission'])
        list_time.append(list_case_clean['formal']['time'])
    elif case == 'ưu tiên xét tuyển':
        list_object.append(list_case_clean['priority']['object'])
        list_register.append(list_case_clean['priority']['register'])
        list_admission.append(list_case_clean['priority']['admission'])
        list_time.append(list_case_clean['priority']['time'])
    elif case == 'đánh giá năng lực':
        list_object.append(list_case_clean['capacity']['object'])
        list_register.append(list_case_clean['capacity']['register'])
        list_admission.append(list_case_clean['capacity']['admission'])
        list_time.append(list_case_clean['capacity']['time'])
```

```python
df_concat['object'] = list_object
df_concat['register'] = list_register
df_concat['admission'] = list_admission
df_concat['time'] = list_time
```
