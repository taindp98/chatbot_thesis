```python
_OLD_DS_2019 = os.path.join(_DATASET,'db_hcmut_mongo_31dec.json')

df_old_ds = load_jsonfile(_OLD_DS_2019)

df_old_ds['year'] = [item[0] for item in list(df_old_ds['year'])]
```

```python
df_ds_2019 = df_old_ds[df_old_ds['year']=='2019'][['major_code','major_name','point','year','subject_group','subject']]

df_ds_2019['major_code'] = [item[0] for item in list(df_ds_2019['major_code'])]

df_ds_2019['major_name'] = [item[0] for item in list(df_ds_2019['major_name'])]

df_ds_2019['point'] = [[float(item[0])] for item in list(df_ds_2019['point'])]

df_ds_2019['year'] = [[item] for item in list(df_ds_2019['year'])]

# df_ds_2019
dict_ds_2019 = df_ds_2019.to_dict('records')
```

```python
_DS_2020 = os.path.join(_DATASET,'database_normalize.json')

df_ds_2020 = load_jsonfile(_DS_2020)

df_ds_2020['major_code'] = [item[0] for item in list(df_ds_2020['major_code'])]

df_ds_2020['major_name'] = [item[0] for item in list(df_ds_2020['major_name'])]

# df_ds_2020.head()

dict_ds_2020 = df_ds_2020.to_dict('records')
```

```python
list_full_feature = list(df_ds_2020.columns)
list_full_feature.remove('major_code')
list_full_feature.remove('major_name')

list_full_feature.remove('point')
list_full_feature.remove('year')

list_full_feature.remove('subject_group')
list_full_feature.remove('subject')
```
