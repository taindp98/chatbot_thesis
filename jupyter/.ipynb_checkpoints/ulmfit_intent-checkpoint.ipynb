## 3 Building a Vocabulary

So instead of jumping directly to the classifier, we could fine-tune our pretrained language model to the corpus and then use that as the base for our classifier.

```python
data_reload=pd.read_csv(os.path.join(data_path,'build_vocab.csv'),encoding='utf-8')
data_reload.columns=['id','content']
data_reload.head()
```

Create databunch

```python
data = (TextList.from_df(data_reload,cols='content')
            .split_by_rand_pct(0.1, seed=42)
            .label_for_lm()           
            .databunch(bs=bs, num_workers=0))
```

```python
data.train_ds
```

* xxbos to indicate the beginning of a text
* xxmaj to indicate the next word was capitalized

## 4 Fituning

```python
df = pd.read_csv(os.path.join(data_path,'train_intent.csv'))
df.loc[pd.isna(df.content),'content']='NA'
df.head()
```

```python
df_train,df_valid = train_test_split(df,stratify=df['label'],test_size=0.3,random_state=42)
```

```python
df_train.shape,df_valid.shape
```

```python
data_lm = TextLMDataBunch.from_df(train_df = df_train, valid_df = df_valid, path = "")
```

```python
data_lm
```

```python
%cd $models_path
```

```python
url_pth='vi_wt'
url_pkl='vi_wt_vocab'
lm_fns=[url_pth,url_pkl]
```

```python
learn_lm = language_model_learner(data_lm, AWD_LSTM, pretrained_fnames=lm_fns, drop_mult=0.5)
```

```python
lr = 1e-1
# lr *= bs/24
```

```python
learn_lm.fit_one_cycle(10,lr, moms=(0.85, 0.65))
```

Find the optimal learning rate

```python
learn_lm.lr_find()
learn_lm.recorder.plot()
```

```python
lr = 1e-1
learn_lm.unfreeze()
learn_lm.fit_one_cycle(8, lr, moms=(0.85,0.65))
```

```python
# learn_lm.save(os.path.join(models_path,'vifine_tuned'))
# learn_lm.save_encoder(os.path.join(models_path,'vifine_tuned_enc'))
```

With the fine-tuning of the language model completed, we build a new databunch for the actual classification problem. The only real difference here is that we supply the actual labels by using "label_from_df" and we pass in a vocab object from the language model training that we performed earlier to make sure they're using same mapping of words to numbers, and then we're ready to create a new "text_classifier_learner", where the library does all the model creation for you behind the scenes.

```python
# bs=21
data_classify = (TextList.from_df(df, vocab=data_lm.vocab, cols='content')
    .split_by_rand_pct(0.1, seed=42)
    .label_from_df(cols='label')
    .databunch(bs=bs, num_workers=0))
```

```python
# data_classify.save(os.path.join(models_path,'data_classify'))
```

```python
# data_classify=load_data(models_path,'data_classify',bs=bs)
```

We load the fine-tuned encoder onto this new model and begin the process of training again

```python
learn_c = text_classifier_learner(data_classify, AWD_LSTM, drop_mult=0.65, metrics=[accuracy])
learn_c.load_encoder('vifine_tuned_enc')
learn_c.freeze()
```

```python
lr=1e-1
lr *= bs/24
```

We could easily improve that by training the language model for more cycles, adding differential learning rates and freezing parts of the model while training.

```python
learn_c.fit_one_cycle(8, lr, moms=(0.85,0.65))
```

```python
learn_c.unfreeze()
learn_c.fit_one_cycle(2, lr, moms=(0.85,0.65))
```

```python
learn_c.freeze_to(-2)
learn_c.fit_one_cycle(2, lr, moms=(0.85,0.65))
```

```python
learn_c.freeze_to(-3)
learn_c.fit_one_cycle(8, slice(lr/2/(2.6**4),lr/2), moms=(0.85,0.65))
```

```python
learn_c.unfreeze()
learn_c.fit_one_cycle(1, lr, moms=(0.85,0.65))
```

```python
learn_c.show_results()
```
