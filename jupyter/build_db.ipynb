```python
import pandas as pd
import json
import os
```

```python
DATA_PATH = '/home/taindp/PycharmProjects/thesis/data'
```

```python
def load_jsonfile(path):
    list_data = []
    for line in open(path, 'r'):
        list_data.append(json.loads(line))
    df = pd.DataFrame.from_dict(list_data)
    return df
```

```python
file_root = os.path.join(DATA_PATH,'db_hcmut_mongo.json')
file_get_rate = os.path.join(DATA_PATH,'db_hcmut_import_mongo.json')
# file_edit = os.path.join(DATA_PATH,'db_hcmut_mongo_edit_16dec.json')
```

```python
db_root = load_jsonfile(file_root)
db_root
```

```python
# delete column
del db_root['university_code']
del db_root['university_name']
del db_root['duration_std']
del db_root['credits']
del db_root['foreign_lang_min']
```

```python
db_get_rate = load_jsonfile(file_get_rate)
dict_get_satify = {}
list_get_satify = []
for item in list(db_get_rate['rate']):
    sent_satify = ['sự hài lòng của sinh viên về chương trình đào tạo là {} trên thang điểm 10'.format(item[0])]
    list_get_satify.append(sent_satify)
dict_get_satify['satify'] = list_get_satify
```
